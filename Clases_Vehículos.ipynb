```python

class MiExcepcion(Exception):
    pass

class Vehiculo:
    
    perdidaPorKm=0.2
    
    def __init__(self, capUsuarios, potMotor, capCombustible, velocidad, kmRecorridos, usuariosCargados, combustible):
        self._capUsuarios=capUsuarios
        self._potMotor=potMotor
        self._capCombustible=capCombustible
        self.velocidad=velocidad
        self.kmRecorridos=kmRecorridos
        self.usuariosCargados=usuariosCargados
        self.combustible=combustible
        
    def __str__(self):
        return(f"capUsuarios: {self.capUsuarios}, potMotor: {self.potMotor}, capCombustible: {self.capCombustible}, velocidad: {self.velocidad}, kmRecorridos: {self.kmRecorridos}, usuariosCargados: {self.usuariosCargados}, combustible: {self.combustible}, ")
        
    @property
    def capUsuarios(self):
        return self._capUsuarios
        
    @property
    def potMotor(self):
        return self._potMotor
        
    @property
    def capCombustible(self):
        return self._capCombustible
        
    @property
    def velocidad(self):
        return self._velocidad
    
    @velocidad.setter
    def velocidad(self, velocidad):
        self._velocidad=velocidad
        
    @property
    def kmRecorridos(self):
        return self._kmRecorridos
    
    @kmRecorridos.setter
    def kmRecorridos(self, kmRecorridos):
        self._kmRecorridos=kmRecorridos
        
    @property
    def usuariosCargados(self):
        return self._usuariosCargados
    
    @usuariosCargados.setter
    def usuariosCargados(self, usuariosCargados):
        self._usuariosCargados=usuariosCargados
        
    @property
    def combustible(self):
        return self._combustible
    
    @combustible.setter
    def combustible(self, combustible):
        self._combustible=combustible
        
    def Acelerar(self, acelerecion):
        if (self.usuariosCargados>0):
            self.velocidad=self.velocidad+acelerecion
        else:
            print("No se puede acelerar si no hay usuarios")
        
    def Desacelerar(self, desaceleracion):
        if (desaceleracion>self.velocidad):
            print(f"No se puede desacelerar {desaceleracion} porque la velocidad actual es {self.velocidad}")
        else:
            self.velocidad-=desaceleracion
        
    def Continuar(self, kms):
        if(kms*self.perdidaPorKm>self.combustible):
            print(f"No se puede avanzar {kms} porque el combustible permite avanzar {kms*self.perdidaPorKm}")
        elif(self.velocidad>0):
            self.combustible-=kms*self.perdidaPorKm
            self.kmRecorridos+=kms
        else:
            print(f"No se puede avanzar {kms} porque la velocidad actual es {self.velocidad}")
            
    def SubirPersonas(self,personas):
        if (self.velocidad>0):
            print(f"No se puede subir {personas} porque el vehículo no está parado")
        elif(self.usuariosCargados+personas>self.capUsuarios):
            print(f"No se puede subir {personas} porque hay {self.usuariosCargados} y el límite es {self.capUsuarios}")
        else:
            self.usuariosCargados+=personas
            
    def Repostar(self,litros):
        if (self.velocidad>0):
            print(f"No se puede repostar {litros} porque el vehículo no está parado")
        elif(self.combustible+litros>self.capCombustible):
            print(f"No se puede repostar {litros} porque hay {self.combustible} y el límite es {self.capCombustible}")
        else:
            self.combustible+=litros
            
class Coche(Vehiculo):
    
    velMax=120
    
    def __init__(self, capUsuarios, potMotor, capCombustible, velocidad, kmRecorridos, usuariosCargados, combustible, cantRuedas, marca, capMaletero):
        super().__init__(capUsuarios, potMotor, capCombustible, velocidad, kmRecorridos, usuariosCargados, combustible)
        self._cantRuedas=cantRuedas
        self._marca=marca
        self._capMaletero=capMaletero
        
    def __str__(self):
            atbs=super().__str__()
            return(f"{atbs} cantRuedas: {self.cantRuedas}, marca: {self.marca}, capMaletero: {self.capMaletero}")
            
    @property
    def cantRuedas(self):
        return self._cantRuedas
        
    @property
    def marca(self):
        return self._marca
        
    @property
    def capMaletero(self):
        return self._capMaletero
    
    def Acelerar(self, aceleracion):
        if (self.velocidad+aceleracion>self.velMax):
            print(f"No se puede acelerar {aceleracion} porque la velocidad maxima es 120")
        else:
            super().Acelerar(aceleracion)
            
class Avion(Vehiculo):
    
    def __init__(self, capUsuarios, potMotor, capCombustible, velocidad, kmRecorridos, usuariosCargados, combustible, conductor, salida, llegada, enVuelo=False):
        super().__init__(capUsuarios, potMotor, capCombustible, velocidad, kmRecorridos, usuariosCargados, combustible)
        self._conductor=conductor
        self._salida=salida
        self._llegada=llegada
        self._enVuelo=enVuelo
        
    def __str__(self):
            atbs=super().__str__()
            return(f"{atbs} conductor: {self.conductor}, salida: {self.salida}, llegada: {self.llegada}, enVuelo: {self.enVuelo}")
            
    @property
    def conductor(self):
        return self._conductor
        
    @property
    def salida(self):
        return self._salida
        
    @property
    def llegada(self):
        return self._llegada
        
    @property
    def enVuelo(self):
        return self._enVuelo
    
    def SubirPersonas(self, personas):
        if (self.enVuelo):
            print("No se puede subir pasajeros porque está en vuelo aún")
        else:
            super().SubirPersonas(personas)
    
    def Repostar(self, personas):
        if (self.enVuelo):
            print("No se puede repostar porque está en vuelo aún")
        else:
            super().Repostar(personas)
            
    def __add__(self,other):
        if (type(other)!=type(self)):
            raise MiExcepcion("Error personalizado: No se puede subir pasajeros porque está en vuelo aún")
        else:
            capUsuarios=self.capUsuarios+other.capUsuarios
            potMotor=self.potMotor+other.potMotor
            capCombustible=self.capCombustible+other.capCombustible
            return(Avion(capUsuarios, potMotor, capCombustible, 0, 0, 0, 0, "Nadie", "nada", "nada", False))
            
def Acelerar(v):
    v.Acelerar(1)
def LeerVehiculo(v):
    print(v)
```
